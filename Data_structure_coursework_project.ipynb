```python
# Afrin Shaikh - DSA253
# Data Structures Assignment

import heapq
from collections import deque

# Q1: Linked List - Playlist Manager
def playlist_task():
    class Song:
        def __init__(self, title, next=None):
            self.title = title
            self.next = next

    class MyPlaylist:
        def __init__(self, first_song):
            self.head = Song("Afrin_" + first_song)
            self.size = 1
            self.limit = 5
            print(f"The playlist is:- Afrin_{first_song}")

        def insert_song(self, title):
            if self.size >= self.limit:
                print("There is a restriction on the playlist. Only five songs can be added.")
                return
            new_song = Song(title)
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_song
            self.size += 1
            print(f"Added: {title}")

        def delete_song(self, title):
            temp, prev = self.head, None
            while temp:
                if temp.title == title:
                    if prev:
                        prev.next = temp.next
                    else:
                        self.head = temp.next
                    self.size -= 1
                    print(f"Deleted: {title}")
                    return
                prev, temp = temp, temp.next
            print("Sorry, Song not found!")

        def show_playlist(self):
            temp = self.head
            print("\nCurrent Playlist")
            i = 1
            while temp:
                print(f"{i}. {temp.title}")
                temp = temp.next
                i += 1
            print(f"Total: {self.size}/5 songs")

    print("\nWelcome to Afrin's Playlist!")
    first_song = input("Enter the first song: ")
    n = MyPlaylist(first_song)
    while True:
        print("\na. Add a Song\nb. Delete a Song\nc. Show the Playlist\nd. END")
        option = input("Option: ")
        if option == 'a':
            n.insert_song(input("Enter song name please: "))
        elif option == 'b':
            n.delete_song(input("Enter song to delete: "))
        elif option == 'c':
            n.show_playlist()
        elif option == 'd':
            break
        else:
            print("Invalid input!")

# Q2: Stack - Sentence Reverser
def reverse_sentence():
    class Node:
        def __init__(self, value):
            self.data = value
            self.next = None

    class Stack:
        def __init__(self):
            self.top = None

        def push(self, value):
            new = Node(value)
            new.next = self.top
            self.top = new

        def pop(self):
            if not self.top:
                return None
            value = self.top.data
            self.top = self.top.next
            return value

    name = input("Enter your name: ").strip()
    tag = name + '_'
    word_count = (len(name) % 5) + 3

    words = []
    print(f"Enter {word_count} words:")
    for i in range(word_count):
        word = input(f"Word {i+1}: ")
        words.append(tag + word)

    print("\nOriginal:", ' '.join(words))

    st = Stack()
    for word in words:
        st.push(word)

    rev = []
    while st.top:
        rev.append(st.pop())

    print("Reversed:", ' '.join(rev))

# Q3: BST - Book Library
def book_bst():
    class Node:
        def __init__(self, title, isbn):
            self.title = title
            self.isbn = isbn
            self.left = None
            self.right = None

    class BST:
        def __init__(self):
            self.root = None

        def insert(self, title, isbn):
            def helper(node):
                if title < node.title:
                    if node.left:
                        helper(node.left)
                    else:
                        node.left = Node(title, isbn)
                else:
                    if node.right:
                        helper(node.right)
                    else:
                        node.right = Node(title, isbn)

            if not self.root:
                self.root = Node("Afrin_" + title, isbn)
            else:
                helper(self.root)

        def inorder(self):
            result = []
            def dfs(node):
                if node:
                    dfs(node.left)
                    result.append((node.title, node.isbn))
                    dfs(node.right)
            dfs(self.root)
            return result

        def search(self, isbn):
            def find(node):
                if not node:
                    return None
                if node.isbn == isbn:
                    return node.title
                return find(node.left) or find(node.right)
            return find(self.root)

    lib = BST()
    books = [
        ("The Science of Everyday Life", "111-1-11111-111-1"),
        ("The Wonders of the Natural World", "222-2-22222-222-2"),
        ("Alice in Wonderland", "333-3-33333-333-3"),
        ("Twisted Love", "444-4-44444-444-4"),
        ("Secret Seven", "555-5-55555-555-5"),
        ("Matilda", "666-6-66666-666-6"),
        ("Malala", "777-7-77777-777-7")
    ]
    books.sort()
    for title, isbn in books:
        lib.insert(title, isbn)

    print("\nMy Library")
    for t, i in lib.inorder():
        print(f"{t} (ISBN: {i})")

    target = input("\nEnter ISBN to search: ")
    result = lib.search(target)
    print("Found:" if result else "Not Found:", result or "Book not in collection")

# Q4: Graph - City Map using BFS
def city_map():
    class Graph:
        def __init__(self):
            self.adj = {}
            self.edges = []

        def add_road(self, u, v, road):
            cost = (ord("A") + ord("f") + ord("r")) % 10
            self.adj.setdefault(u, []).append((v, road, cost))
            self.adj.setdefault(v, []).append((u, road, cost))
            self.edges.append((u, v, road, cost))

        def bfs(self, start):
            visited = set()
            q = deque([start])
            visited.add(start)
            print("\nBFS from", start)
            while q:
                city = q.popleft()
                print(city, end=" → ")
                for neigh, _, _ in self.adj[city]:
                    if neigh not in visited:
                        visited.add(neigh)
                        q.append(neigh)
            print("End")

        def show_roads(self):
            print("\nRoads")
            for u, v, road, cost in self.edges:
                print(f"{u} -> {v} via {road} (Cost: {cost})")

    g = Graph()
    g.add_road("Afrin_Pune", "Kolkata", "Chennai")
    g.add_road("Kolkata", "Bangalore", "Solapur")
    g.add_road("Afrin_Pune", "Bangalore", "Latur")
    g.add_road("Bangalore", "Chennai", "Mysore")

    g.bfs("Afrin_Pune")
    g.show_roads()

# Q5: Min-Heap - Student Score Manager
def student_scores():
    name = "Afrin"
    count = (len(name) % 5) + 5
    marks = [34.7, 78, 45, 99, 80][:count]
    heap = marks[:]
    heapq.heapify(heap)

    print("\nScore Manager")
    print("Initial Heap:", heap)

    while True:
        print("\nMenu:")
        print("a. Show Min")
        print("b. Remove Min")
        print("c. Show Updated Heap")
        print("d. End")
        option = input("Choice: ")

        if option == 'a':
            print("Minimum Score:", heap[0] if heap else "Empty")
        elif option == 'b':
            print("Removed:", heapq.heappop(heap) if heap else "Heap empty")
        elif option == 'c':
            print("Updated Heap:", heap)
        elif option == 'd':
            break
        else:
            print("Invalid option")

# MAIN MENU
def main():
    while True:
        print("\nAfrin's Data Structure Assignment (DSA253)")
        print("\nChoose an option")
        print("a. Playlist Manager\nb. Sentence Reverser\nc. Book Library\nd. City Map\ne. Student Scores\nf. End")
        c = input("Choose the option: ")
        if c == 'a':
            playlist_task()
        elif c == 'b':
            reverse_sentence()
        elif c == 'c':
            book_bst()
        elif c == 'd':
            city_map()
        elif c == 'e':
            student_scores()
        elif c == 'f':
            print("Completed!")
            break
        else:
            print("Please select a valid option (a-f)")

if __name__ == '__main__':
    main()
```
